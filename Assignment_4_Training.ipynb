### **Q6 [Boolean Filter] Filter students where Score > 80. How many students scored above 80?**

```python
high_score_students = df[df['Score'] > 80]
display(high_score_students)
print(f"Number of students who scored above 80: {len(high_score_students)}")
```

### **Q7 [Boolean Filter] Filter students from Delhi who are in CSE department.**

```python
delhi_cse_students = df[(df['City'] == 'Delhi') & (df['Department'] == 'CSE')]
display(delhi_cse_students)
```

```python
delhi_cse_students_selected_cols = delhi_cse_students[['Name', 'Score', 'Grade']]
display(delhi_cse_students_selected_cols)
```

### **Q8 [Boolean Filter] Find all students who Failed (Passed == False). What is their average Score?**

```python
failed_students = df[df['Passed'] == False]
display(failed_students)

average_score_failed = failed_students['Score'].mean()
print(f"Average score of students who failed: {average_score_failed:.2f}")
```

### **Q9 [String Filter] Filter students whose Name contains the string 'a' (case-insensitive) using .str.contains().**

```python
students_with_a_in_name = df[df['Name'].str.contains('a', case=False, na=False)]
display(students_with_a_in_name)
```

### **Q10 [Add Column] Add a new column 'Score_Category': if Score >= 80 → 'Distinction', Score >= 60 → 'Merit', else 'Pass/Fail'. Use np.where() or nested condition.**

```python
df['Score_Category'] = np.where(df['Score'] >= 80, 'Distinction',
                                np.where(df['Score'] >= 60, 'Merit', 'Pass/Fail'))
display(df.head()) # Display the first few rows to show the new column
```

### **Sort the DataFrame by Score in descending order.**

```python
df_sorted = df.sort_values(by='Score', ascending=False)
display(df_sorted.head())
```

### **Reset the index.**
